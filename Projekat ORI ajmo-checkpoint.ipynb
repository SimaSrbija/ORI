```python
#tukey interquartile range
def find_outliers_tukey(x):
    q1 = np.percentile(x,25)
    q3 = np.percentile(x,75)
    iqr = q3 - q1
    floor = q1 - 1.5*iqr
    celing = q3 + 1.5*iqr
    outlier_indices = list(x.index[(x < floor) | (x > celing)])
    outlier_values = list(x[outlier_indices])
    
    return outlier_indices, outlier_values
```

```python
def find_outliers_tukey2(x):
    q1 = np.percentile(x,25)
    q3 = np.percentile(x,75)
    iqr = q3 - q1
    floor = q1 - 8*iqr
    celing = q3 + 8*iqr
    outlier_indices = list(x.index[(x < floor) | (x > celing)])
    outlier_values = list(x[outlier_indices])
    
    return outlier_indices, outlier_values
```

```python
#kada izuzmemo ekstremne vrednosti yearpublished, definitivno je bolje mozes probati posle to sa average_rating i jos necim ali sada idemo dalje
tukey_indices, tukey_values = find_outliers_tukey(games["yearpublished"])
lista_vr = np.sort(tukey_values)
print(lista_vr)
lista_ind = np.sort(tukey_indices)
print(lista_ind)
print(len(tukey_values))
```

```python
games = games.loc[~games.index.isin(tukey_indices)]
```

```python
#kada izuzmemo ekstremne vrednosti yearpublished, definitivno je bolje mozes probati posle to sa average_rating i jos necim ali sada idemo dalje
tukey_indices, tukey_values = find_outliers_tukey2(games["total_wanters"])
lista_vr = np.sort(tukey_values)
print(lista_vr)
lista_ind = np.sort(tukey_indices)
print(lista_ind)
print(len(tukey_values))
```

```python
X = games.drop("average_rating", axis = 1)
y = games.average_rating
#X = X.drop("bayes_average_rating", axis = 1)
X = X.drop("meta", axis=1)
X = X.drop("id", axis=1)
X = X.drop("type", axis = 1)
X = X.drop("name", axis = 1)

print(X.dtypes)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
Linear = LinearRegression()
Linear.fit(X_train, y_train)
```

```python
linearPrediction = Linear.predict(X_test)
```

```python
from sklearn import metrics
print('LR Mean Squared Error:', metrics.mean_squared_error(y_test, linearPrediction))
print('LR Mean Absolute Error:', metrics.mean_absolute_error(y_test, linearPrediction))  
print('LR Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, linearPrediction)))
```

```python
import pandas as pd  
dfLinear = pd.DataFrame({"Actual value": y_test, "Predicted value": linearPrediction})
dfLinear
```

```python
dfLinear1 = dfLinear.head(10)
dfLinear1.plot(kind = "bar", figsize = (18,10))
plt.grid(linewidth="0")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
RandomForest = RandomForestRegressor(n_estimators = 100, max_depth=None, min_samples_leaf = 5, random_state = 20)
RandomForest.fit(X_train, y_train)
```

```python
RandomForestPrediction = RandomForest.predict(X_test)

print("RF Forest Mean Squared Error:", metrics.mean_squared_error(y_test, RandomForestPrediction))
print('RF Mean Absolute Error:', metrics.mean_absolute_error(y_test, RandomForestPrediction))  
print('RF Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, RandomForestPrediction)))
```
